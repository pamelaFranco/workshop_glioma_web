# **Física de la Resonancia Magnética:**

# **Ecuaciones de Bloch**

La evolución de la magnetización tras un pulso de radiofrecuencia (RF) de ángulo $\alpha$ se rige por los procesos de relajación térmica y pérdida de fase.

La evolución de la magnetización $\vec{M}$ en un campo magnético $\vec{B}$ se describe mediante la ecuación de movimiento vectorial, que relaciona el torque ejercido por el campo sobre el momento magnético:

$$\frac{d\vec{M}}{dt} = \gamma (\vec{M} \times \vec{B}) + \text{Términos de Relajación}$$

Donde $\gamma$ es la constante giromagnética. Si descomponemos esta ecuación vectorial en sus componentes cartesianas, obtenemos las ecuaciones diferenciales que rigen la relajación.


# **1. Magnetización Longitudinal ($M_z$)**

La magnetización en el eje $z$ tiende a volver a su estado de equilibrio ($M_0$) a través de la interacción con la red (entorno). La ecuación diferencial es:

$$\frac{dM_z(t)}{dt} = \frac{M_0 - M_z(t)}{T_1}$$

**Resolución:**
Si aplicamos integración por separación de variables desde un tiempo $0$ hasta $t$:$$\int_{M_z(0)}^{M_z(t)} \frac{dM_z}{M_0 - M_z} = \int_{0}^{t} \frac{dt}{T_1}$$La solución resultante es la que utilizamos para calcular el contraste en el simulador:$$M_z(t) = M_0 \cdot (1 - e^{-t/T_1}) + M_z(0) e^{-t/T_1}$$

En nuestra simulación:
* $M_0$: Es la magnetización en equilibrio, proporcional a la densidad protónica ($\rho$).
* $M_z(0)$: Es la magnetización inmediatamente después del pulso de RF, calculada como $M_{ss} \cdot \cos(\alpha)$, donde $M_{ss}$ es el estado estacionario.

# **2. Magnetización Transversal ($M_{xy}$)**
Tras un pulso de radiofrecuencia, los espines pierden coherencia de fase debido a interacciones espín-espín. El cambio de la magnetización transversal en el tiempo es:
$$\frac{dM_{xy}(t)}{dt} = -\frac{M_{xy}(t)}{T_2}$$

**Resolución:**

Esta es una ecuación diferencial de decaimiento exponencial simple:$$\frac{dM_{xy}}{M_{xy}} = -\frac{1}{T_2} dt \implies \ln(M_{xy}) = -\frac{t}{T_2} + C$$Al resolver para las condiciones iniciales, obtenemos el decaimiento que define la señal detectable:$$M_{xy}(t) = M_{xy}(0) e^{-t/T_2}$$

* **Spin Echo (SE):** El decaimiento depende de $T_2$ puro, ya que los pulsos de reenfase eliminan las inhomogeneidades del campo.

* **Gradient Echo (GE):** En secuencias de GE, sustituimos $T_2$ por $T_2^*$, donde $\frac{1}{T_2^*} = \frac{1}{T_2} + \gamma \Delta B_0$, incluyendo así el efecto de las inhomogeneidades del campo magnético. El decaimiento es más rápido y depende de $T_2^*$, que incluye las inhomogeneidades del campo magnético.

# **3. Inversion Recovery (IR)**

En una secuencia IR, el objetivo es alcanzar un equilibrio dinámico tras aplicar repetidamente pulsos de 180° separados por un tiempo $TR$. Para resolverlo, analizamos la magnetización longitudinal $M_z$ en puntos clave del ciclo.1.

**Ecuación de Relajación Longitudinal**

Como vimos anteriormente, la solución general de la ecuación diferencial de Bloch para $M_z$ es:
$$M_z(t) = M_0 - (M_0 - M_z(0))e^{-t/T_1}$$

**Condiciones de Contorno en Equilibrio (Steady State)**

Definimos dos momentos críticos en el ciclo:

* $M_z^-$: Magnetización justo antes del pulso de 180°.
* $M_z^+$: Magnetización justo después del pulso de 180°. Debido al pulso inversor, $M_z^+ = -M_z^-$.

En estado estacionario, la magnetización debe recuperarse durante el tiempo $TR$ para volver exactamente al mismo valor antes del siguiente pulso:$$M_z(TR) = M_z^-$$

**Resolución:**
Sustituimos las condiciones en la solución general evaluada en $t = TR$:$$M_z^- = M_0 - (M_0 - M_z^+)e^{-TR/T_1}$$Sustituyendo $M_z^+ = -M_z^-$:$$M_z^- = M_0 - (M_0 - (-M_z^-))e^{-TR/T_1}$$$$M_z^- = M_0 - (M_0 + M_z^-)e^{-TR/T_1}$$Despejamos $M_z^-$:$$M_z^- (1 + e^{-TR/T_1}) = M_0 (1 - e^{-TR/T_1})$$$$M_z^- = M_0 \frac{1 - e^{-TR/T_1}}{1 + e^{-TR/T_1}}$$

**Magnetización en el Tiempo de Inversión ($TI$)**

La señal que realmente medimos ocurre en el tiempo $TI$, partiendo desde $M_z^+$:$$M_z(TI) = M_0 - (M_0 - M_z^+)e^{-TI/T_1}$$$$M_z(TI) = M_0 - (M_0 + M_z^-)e^{-TI/T_1}$$Sustituyendo el valor de $M_z^-$ encontrado arriba, llegamos a la ecuación final del estado estacionario para IR:$$M_z(TI) = M_0 \left( 1 - \frac{2e^{-TI/T_1}}{1 + e^{-TR/T_1}} \right)$$

**La Aproximación de Inversion Recovery (IR)**

En la práctica clínica, a menudo se utiliza una versión simplificada de la ecuación de Bloch para IR. Esta aproximación asume que el Tiempo de Repetición ($TR$) es lo suficientemente largo como para que la magnetización longitudinal se recupere por completo ($M_z \approx M_0$) antes de cada nuevo pulso de inversión.

**La Ecuación Simplificada**

Si asumimos que $TR \gg T_1$, entonces el término $e^{-TR/T1} \approx 0$. Al aplicar esto a la solución del estado estacionario:$$M_z(TI) = M_0 \left( 1 - \frac{2e^{-TI/T_1}}{1 + \underbrace{e^{-TR/T_1}}_{\approx 0}} \right) \implies M_z(TI) \approx M_0 (1 - 2e^{-TI/T_1})$$Esta es la fórmula que la mayoría de los usuarios utilizan para calcular el punto de anulación de un tejido.

**El Punto de Anulación (Null Point)**

El objetivo de secuencias como STIR (para anular grasa) o FLAIR (para anular líquido cefalorraquídeo) es encontrar el tiempo $TI$ exacto donde la curva cruza el eje cero ($M_z = 0$).Usando la aproximación, despejamos $TI$:$$0 = M_0 (1 - 2e^{-TI/T_1})$$$$2e^{-TI/T_1} = 1$$$$TI_{null} = T_1 \cdot \ln(2) \approx 0.693 \cdot T_1$$

**Comparación: ¿Cuándo falla la aproximación?**

Es importante que en el laboratorio los estudiantes noten que:

* **Con $TR$ largo:** La aproximación es excelente y el contraste es máximo.

* **Con $TR$ corto:** La magnetización no llega a $M_0$ antes de la inversión. Esto reduce el contraste total y desplaza ligeramente el punto de anulación real, requiriendo la fórmula completa del estado estacionario.

### **El Ángulo de Ernst**
En GE, existe un ángulo óptimo llamado **Ángulo de Ernst** que maximiza la señal para un $TR$ y $T_1$ dados:

$$\cos(\alpha_{Ernst}) = e^{-TR/T_1}$$

```python
# @title Simulador de RM: Interfaz Profesional { display-mode: "form" }
# @markdown <-- Haz clic en Play para activar la aplicación.

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Contenedor de salida para los gráficos
out = widgets.Output()

def calcular_mri(tr, te, alpha, et, ir, ti_sel, t1a, t2a, t2sa, ra, t1b, t2b, t2sb, rb):
    t = np.linspace(0, tr, 500)
    t2 = np.linspace(0, te + 40, 500)

    # Lógica de Magnetización Longitudinal (Mz)
    if ir == "On":
        if ti_sel == "Null A":
            ti = -t1a * np.log((1 + np.exp(-tr/t1a)) / 2)
        elif ti_sel == "Null B":
            ti = -t1b * np.log((1 + np.exp(-tr/t1b)) / 2)
        else: ti = 500
        ma_val = ra * (1 - 2*np.exp(-ti/t1a) + np.exp(-tr/t1a))
        mb_val = rb * (1 - 2*np.exp(-ti/t1b) + np.exp(-tr/t1b))
    else:
        ma_val = ra * (1 - np.exp(-tr/t1a)) / (1 - np.cos(np.radians(alpha)) * np.exp(-tr/t1a))
        mb_val = rb * (1 - np.exp(-tr/t1b)) / (1 - np.cos(np.radians(alpha)) * np.exp(-tr/t1b))

    da, db = (t2sa, t2sb) if et == 'GE' else (t2a, t2b)
    ia = abs(ma_val * np.sin(np.radians(alpha)) * np.exp(-te/da))
    ib = abs(mb_val * np.sin(np.radians(alpha)) * np.exp(-te/db))

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5), facecolor='white')

    # Plot Mz
    mza = ra * (1 - np.exp(-t/t1a)) + (ma_val * np.cos(np.radians(alpha))) * np.exp(-t/t1a)
    mzb = rb * (1 - np.exp(-t/t1b)) + (mb_val * np.cos(np.radians(alpha))) * np.exp(-t/t1b)
    ax1.plot(t, mza, 'tab:blue', label='Mat. A', lw=2)
    ax1.plot(t, mzb, 'tab:red', label='Mat. B', lw=2)
    ax1.set_title('Magnetización Longitudinal $M_z$'); ax1.set_xlabel('Tiempo (ms)'); ax1.set_ylabel('Intensidad (u.a.)'); ax1.grid(True, alpha=0.3); ax1.legend()

    # Plot Mxy
    ax2.plot(t2, abs(ma_val * np.sin(np.radians(alpha))) * np.exp(-t2/da), 'tab:blue', lw=2)
    ax2.plot(t2, abs(mb_val * np.sin(np.radians(alpha))) * np.exp(-t2/db), 'tab:red', lw=2)
    ax2.axvline(te, color='green', ls='--', label=f'TE={te}ms')
    ax2.set_title('Magnetización Transversal $M_{xy}$'); ax2.set_xlabel('Tiempo (ms)'); ax2.set_ylabel('Intensidad (u.a.)'); ax2.grid(True, alpha=0.3); ax2.legend()

    # Fantoma
    size = 120
    xx, yy = np.meshgrid(np.linspace(-1, 1, size), np.linspace(-1, 1, size))
    d = np.sqrt(xx**2 + yy**2)
    ph = np.zeros((size, size))
    ph[d < 0.8], ph[d < 0.4] = ib, ia
    im = ax3.imshow(ph + np.random.normal(0, 0.01, ph.shape), cmap='gray', vmin=0, vmax=1.2)

    # Leyendas en el fantoma
    ax3.text(0, 10, 'Mat. A', color='blue', fontweight='bold')
    ax3.text(0, 25, 'Mat. B', color='red', fontweight='bold')
    ax3.set_title(f'Imagen RM (ΔI: {abs(ia-ib):.3f})')
    ax3.axis('off')
    plt.colorbar(im, ax=ax3, label='Intensidad (u.a.)')
    plt.show()

def runner(change=None):
    with out:
        clear_output(wait=True)
        calcular_mri(s_tr.value, s_te.value, s_alpha.value, d_et.value, d_ir.value, d_ti.value,
                     s_t1a.value, s_t2a.value, s_t2sa.value, s_ra.value,
                     s_t1b.value, s_t2b.value, s_t2sb.value, s_rb.value)

# --- Widgets Declarados Correctamente ---
style = {'description_width': 'initial'}

# Secuencia
s_tr = widgets.IntSlider(value=600, min=10, max=4000, step=50, description='TR (ms)', style=style)
s_te = widgets.IntSlider(value=15, min=1, max=300, step=2, description='TE (ms)', style=style)
s_alpha = widgets.IntSlider(value=90, min=0, max=180, step=5, description='Alpha (°)', style=style)
d_et = widgets.Dropdown(options=['SE', 'GE'], value='SE', description='Secuencia', style=style)
d_ir = widgets.Dropdown(options=['Off', 'On'], value='Off', description='IR Mode', style=style)
d_ti = widgets.Dropdown(options=['Manual', 'Null A', 'Null B'], value='Manual', description='TI Sel.', style=style)

# Material A
s_t1a = widgets.IntSlider(value=600, min=100, max=4000, description='T1 (ms)', style=style)
s_t2a = widgets.IntSlider(value=80, min=10, max=1000, description='T2 (ms)', style=style)
s_t2sa = widgets.IntSlider(value=40, min=5, max=200, description='T2* (ms)', style=style)
s_ra = widgets.FloatSlider(value=0.7, min=0.1, max=2.0, step=0.1, description='rho (ρ)', style=style)

# Material B
s_t1b = widgets.IntSlider(value=2500, min=100, max=4000, description='T1 (ms)', style=style)
s_t2b = widgets.IntSlider(value=500, min=10, max=1000, description='T2 (ms)', style=style)
s_t2sb = widgets.IntSlider(value=80, min=5, max=200, description='T2* (ms)', style=style)
s_rb = widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='rho (ρ)', style=style)

# Observadores
widgets_list = [s_tr, s_te, s_alpha, d_et, d_ir, d_ti, s_t1a, s_t2a, s_t2sa, s_ra, s_t1b, s_t2b, s_t2sb, s_rb]
for w in widgets_list:
    w.observe(runner, names='value')

# Layout de la UI
ui = widgets.VBox([
    widgets.HTML("<b>Configuración de Secuencia</b>"),
    widgets.HBox([s_tr, s_te, s_alpha]),
    widgets.HBox([d_et, d_ir, d_ti]),
    widgets.HTML("<b>Material A (Centro)</b>"),
    widgets.HBox([s_t1a, s_t2a, s_t2sa, s_ra]),
    widgets.HTML("<b>Material B (Periferia)</b>"),
    widgets.HBox([s_t1b, s_t2b, s_t2sb, s_rb])
])

display(ui, out)
runner()
```
